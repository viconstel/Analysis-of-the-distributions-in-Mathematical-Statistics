# Задача 2

```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd
import seaborn as sns
%matplotlib inline
DATA_PATH = 'C:\\repos\Matstat\\'
r_1, r_2 = [], []
with open(f'{DATA_PATH}S2_02_001.dat') as f:
    for i in f:
        a, b = i.split()
        a, b = float(a), float(b)
        r_1.append(a)
        r_2.append(b)
plt.plot(r_1,r_2,'.')
plt.show()
sns.jointplot(x=pd.Series(r_1), y=pd.Series(r_2))
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd
import seaborn as sns
import math
%matplotlib inline

DATA_PATH = 'C:\\repos\Matstat\\' 
with open(f'{DATA_PATH}S2_04_002.dat') as f: 
    data = [float(i.strip()) for i in f] 
mas=np.sort(data)
lenth=len(data) 
x1=[]
i=0
while i<lenth:
    x1.append(((i+1)/lenth))
    i=i+1
x1.reverse()
plt.plot(mas,x1,'bo')
x=np.arange(0,7,0.01)
plt.plot(x,1/x**2,'r')
plt.yscale('log')
plt.xscale('log')
plt.axis([1,5,0.01,1])
plt.show()
```
